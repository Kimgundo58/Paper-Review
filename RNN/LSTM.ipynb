## Abstract

LSTM은 기존 RNN의 **기울기 소실 문제(Vanishing Gradient)**를 해결하기 위해 제안된 모델입니다.

---

## Previous Study — 기존 RNN의 문제

RNN은 시퀀스 데이터를 처리할 수 있지만, 역전파 시 gradient가 소실됩니다.

$$\frac{\partial L}{\partial W} = \sum_{t} \frac{\partial L_t}{\partial W}$$

시간이 길어질수록 앞쪽 시점의 gradient가 0에 수렴합니다.

---

## LSTM 구조

### Forget Gate
이전 정보를 얼마나 잊을지 결정합니다.

$$f_t = \sigma(W_f \cdot [h_{t-1}, x_t] + b_f)$$

### Input Gate
새로운 정보를 얼마나 기억할지 결정합니다.

$$i_t = \sigma(W_i \cdot [h_{t-1}, x_t] + b_i)$$

### Output Gate

$$o_t = \sigma(W_o \cdot [h_{t-1}, x_t] + b_o)$$

---

## 코드 구현

```python
import torch
import torch.nn as nn

# PyTorch 내장 LSTM
lstm = nn.LSTM(input_size=10, hidden_size=20, num_layers=1, batch_first=True)

x = torch.randn(8, 5, 10)  # (batch, seq_len, input_size)
output, (h_n, c_n) = lstm(x)

print(f"output shape: {output.shape}")   # (8, 5, 20)
print(f"hidden shape: {h_n.shape}")      # (1, 8, 20)
print(f"cell   shape: {c_n.shape}")      # (1, 8, 20)
```
